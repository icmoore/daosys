# Brownian class
In the Python code below, we define a class Brownian with a few useful methods,

* **gen_gbm()**: Generates motion from the Geometric Brownian Motion process
* **gen_normal()**: Generates motion by drawing from the Normal Distribution
* **gen_random_walk()**: Generates motion from the Random Walk process

```python
import numpy as np
import matplotlib.pyplot as plt

import os
cwd =  os.getcwd().replace("notebooks/research/impermanent_loss","")
os.chdir(cwd)

from python.dev.math.model import BrownianModel
```

```python
b = BrownianModel(0.2)
paths = b.gen_gbms(0.1, 1, 365, 1)

fig, (gbm_ax) = plt.subplots(nrows=1, sharex=True, sharey=False, figsize=(12, 5))

for i in range(paths.shape[1]):
    gbm_ax.plot(paths[:,i])
plt.show()
```
