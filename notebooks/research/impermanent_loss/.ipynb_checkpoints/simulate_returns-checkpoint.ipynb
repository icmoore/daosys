**Calculations**

```python
N = 1000

sys_arr = simLiq.get_x_arr()
dai_arr = simLiq.get_y_arr()

lpVal = LiquidityPoolUSD(sys_arr, dai_arr, p_arr)
hodl_usd = lpVal.hodl_position_usd(sys_arr[0], dai_arr[0])
lp_pos = lpVal.lp_position_usd(sys_arr[0], dai_arr[0])
lp_mc_usd = lpVal.lp_marketcap_usd()
lp_price = lpVal.usd_per_lp(N)
```

```python
fee_sys_arr = simLiq.get_fee_x_arr()
fee_dai_arr = simLiq.get_fee_y_arr()
fees = fee_dai_arr + p_arr[1:]*fee_sys_arr
```

```python
init_pos_sys = 1000
init_pos_dai = p_arr[0]*init_pos_sys
N = 1000

lpVal2 = LiquidityPoolUSD(sys_arr, dai_arr, p_arr)
lp_pos = lpVal2.lp_position_usd(init_pos_sys, init_pos_dai)
lp_tot = lpVal2.lp_tot_usd()

collected_fees = fees[N:]*lp_pos[(N+1):]/lp_tot[(N+1):]
lp_pos_collected_fees = np.cumsum(collected_fees)

percent_returns = lp_pos_collected_fees/lp_pos[N]

print('percent returns: {:.3f}%'.format(100*percent_returns[-1]))
print('collected fees:  ${:.2f}'.format(lp_pos_collected_fees[-1]))
```

```python
class LPSimFees():
    
    def __init__(self, simLiq, lpVal):
        self.__sim_liq = simLiq
        self.__lp_val = lpVal
        
    def tot_fees_usd(self)
        p_arr = simLiq.get_price_arr
        fee_sys_arr = simLiq.get_fee_x_arr()
        fee_dai_arr = simLiq.get_fee_y_arr()
        fees = fee_dai_arr + p_arr[1:]*fee_sys_arr
        return fees
        
    def position_fees_usd(self, pos_x, pos_y, N = 0)
        fees = self.tot_fees_usd()
        lp_pos = self.__lp_val.lp_position_usd(pos_x, pos_y)
        lp_tot = self.__lp_val.lp_tot_usd()
        collected_fees = fees[N:]*lp_pos[(N+1):]/lp_tot[(N+1):]
        return collected_fees
    
    def cum_position_fees_usd(self, pos_x, pos_y, N = 0)
        collected_fees = self.collected_fees_usd(pos_x, pos_y, N = 0)
        return np.cumsum(collected_fees)
    
        
        
```
